Este notebook documenta el proceso de carga, limpieza y analisis estadistico del set de datos de peliculas.

Importación de las librerias necesarias para realizar el analisis.

En la siguiente línea de código se importan las clases CargadorDatos y ProcesadorEDA desde el archivo de código ClaseProcesadorEDA del paquete eda. También se importa la librería pandas.

```python
from eda.ClaseProcesadorEDA import CargadorDatos, ProcesadorEDA
import pandas as pd
```

Carga del set de datos y Exploración de los datos

En la siguente línea de código se define la ruta del archivo .CSV, ubicado en la carpeta "raw" dentro de "data".
Así se utiliza clase CargadorDatos para cargar el archivo y obtener el dataframe original.

Luego se realiza la inizalición de la clase ProcesadorEDA con el set de datos cargado y se realiza la exploración del conjunto de datos.

```python
ruta = "../data/raw/tmdb_2020_to_2025.csv"
cargador = CargadorDatos(ruta)
df_original = cargador.cargar_csv()

procesador = ProcesadorEDA(df_original)
procesador.exploracion()
```

Limpieza del set de datos

En la siguiente línea de código se aplica la limpieza del set de datos utilizando la clase ProcesadorEDA. Luego se confirma que la limpieza se haya realizado correctamente y se muestra el set de datos limpio.

```python
df_limpio = procesador.limpieza_datos()
print("Los datos han sido limpiados exitosamente")
print(df_limpio)
```

Visualizadores del resumen estadistico y la matriz de correlación

En la siguiente línea de código se aplica los métodos resumen estadistico y la matriz de correlación. Permitiendo mostrar el resumen estadistico del set de datos limpio y la matriz de correlacion para analizar las relaciones entre las variables numericas.

```python
#Muestra el resumen de correlacion
display(procesador.resumen_estadistico())

#Muestra la matriz de correlacion
display(procesador.matriz_correlacion())
```
